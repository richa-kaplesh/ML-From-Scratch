### Magnitude of a vector

```python
##2D
v1 = np.array([1,2])
value = v1[0]**2 +v1[1]**2
magnitude = np.sqrt(value)
print(f"magnitude of {v1} is {magnitude}")
```

```python
##3D
##manual way for intuition
v2 = np.array([2,3,2])
projected_v2= v2[0]**2+v2[2]**2
projected = np.sqrt(projected_v2)
magnitude_v2 = projected**2+v2[1]**2
real_magnitude = np.sqrt(magnitude_v2)
print(f"first projected xz {projected} then second pythagoras {real_magnitude}")
```

```python
### faster way
v2 = np.array([2,3,2])
projected_v2= v2[0]**2+v2[1]**2+v2[2]**2
projected = np.sqrt(projected_v2)
print(f"{projected:.3f}")
```

### Normalize vector 
Resulting in a unit vector with a length of 1 that points in the same direction. 
We first basically use euclidean theorem to i.r sum of squares of values in vector then their squareroot(i) , then to normalise it we divide each value in vector with i  

```python
v3 = np.array([3,4])
summie=  sum(a ** 2 for a in v3)
squariee = np.sqrt(summie)
normiee = [a/squariee for a in v3]
normiee = v3/squariee
normiee
```

Which formula are you asking about?

**Option 1 — Inverse formula for 2×2:**

$$A^{-1} = \frac{1}{ad-bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$$

You already coded this!

---

**Option 2 — The identity relationship:**

$$A \times A^{-1} = I$$

---

**Option 3 — How to find inverse using determinant:**

$$\text{if } \det(A) \neq 0 \rightarrow A^{-1} \text{ exists}$$
$$\text{if } \det(A) = 0 \rightarrow A^{-1} \text{ does NOT exist}$$

---

**Option 4 — For finding what input gave an output:**

$$A \times x = b$$
$$x = A^{-1} \times b$$

This is actually the most useful one in practice — used to **solve systems of equations!**
